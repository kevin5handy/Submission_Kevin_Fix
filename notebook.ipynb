# Cleaning Data

Dilakukan pengubahan tipe data 'dteday' yang awalnya bertipe 'object' menjadi 'datetime' pada kedua dataset.

```python
df1['dteday'] = pd.to_datetime(df1['dteday'])
df1.info()
```

```python
df2['dteday'] = pd.to_datetime(df2['dteday'])
df2.info()
```

# Exploratory Data Analysis

Pertanyaan bisnis yang ingin diteliti yaitu sebagai berikut.
1. Bagaimana progres jumlah pengguna sepeda (casual, registered, count) setiap bulannya?
2. Bagaimana pengaruh cuaca terhadap jumlah pelanggan sepeda?
3. Perhari, kapankah waktu dimana didapat jumlah pelanggan sepeda terbanyak?

> Pertanyaan 1

```python
df1.describe(include="all")
```

```python
df1.corr()
```

```python
mdf1 = df1.resample(rule='M', on='dteday').agg({
    "cnt": "sum"
})

mdf1.index = mdf1.index.strftime('%Y-%m')
mdf1 = mdf1.reset_index()
mdf1.rename(columns={
    "dteday": "month",
    "cnt": "cnt_monthly"
}, inplace=True)
mdf1
```

Tabel di atas mencakup detail dari jumlah keseluruhan pengguna sepeda setiap bulannya pada tahun 2011 dan 2012. Terlihat bahwa jumlah tertinggi yaitu pada 2012-9 dan terendahnya yaitu pada 2011-01.

> Pertanyaan 2

```python
wdf1 = df1.groupby(by=["weathersit"]).agg({
    "casual": "sum",
    "registered": "sum",
    "cnt": "sum"
})

wdf1 = wdf1.reset_index()
wdf1.rename(columns={
    "casual": "casual_w",
    "registered": "registered_w",
    "cnt": "cnt_w"
}, inplace=True)
wdf1
```

```python
wdf1["weather"] = wdf1["weathersit"].apply(lambda x: "clear" if x == 1 else "mist" if x == 2 else "light snow" if x == 3 else "heavy rain")
wdf1
```

Tabel di atas mencakup hasil penghitungan jumlah pengguna sepeda (tipe casual, registered, dan count) untuk berbagai kondisi cuaca. Dapat dilihat bahwa cuaca meberikan pengaruh yang signifikan terdahap seluruh tipe pengguna sepeda. Semakin cerah cuaca, semakin banyak yang menggunakan sepeda.

> Pertanyaan 3

```python
df2.describe(include="all")
```

```python
df2.corr()
```

```python
hdf2 = df2.groupby(by=["hr"]).agg({
    "cnt": "mean"
})

hdf2
```

Tabel di atas mencakup informasi rata-rata pengguna sepeda pada tiap jam dalam sehari. Dapat dilihat bahwa jumlah pengguna sepeda relatif tinggi pada jam 8 pagi dan pada sore hari dengan rentang pukul 16-19. Sedangan jumlah pengguna yang relatif paling sedikit yaitupada dini hari pada pukul 1-5.

# Data Visualization

> Pertanyaan 1

```python
plt.figure(figsize=(12,6))
plt.plot(mdf1["month"], mdf1["cnt_monthly"], marker='o', linewidth=2, color="#72BCD4")
plt.title("Judul", loc="center", fontsize = 10)
plt.xticks(fontsize = 11, rotation=45)
plt.yticks(fontsize = 11)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(12,6))

sns.barplot(data=wdf1, x="weather", y="casual_w", ax=ax[0])
ax[0].set_title("Casual Weather", loc="center")
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].tick_params(axis='x', labelsize=11)
sns.barplot(data=wdf1, x="weather", y="registered_w", color='red', ax=ax[1])
ax[1].set_title("Registered Weather", loc="center")
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].tick_params(axis='x', labelsize=11)
sns.barplot(data=wdf1, x="weather", y="cnt_w", color='green',ax=ax[2])
ax[2].set_title("Count Weather", loc="center")
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].tick_params(axis='x', labelsize=11)
plt.show()
```
